```python

def parse_xls(file_url,rows_to_skip,reject_value):
    book = load_workbook(file_url)
    sheet = book.active
    rows = sheet.rows
    # Skip any rows at the top
    while rows_to_skip > 0:
        next(rows)
        rows_to_skip -= 1
    headers = [cell.value for cell in next(rows)]
    skip_row = False
    data_dict = {}
    data_rows = []
    final_data_set = []
    for row in rows:
        cell_record = []
        for cell in row:
            cell_data = cell.value
            if str(cell_data).rstrip() == reject_value or skip_row is True :
                skip_row = False
                continue
            else:
                cell_record.append(cell.value)
        data_rows.append(cell_record)
    
    for data_row in data_rows:
        if data_row[0] is None:
            continue
        else:
            data_dict = dict(zip(headers,data_row))
        final_data_set.append(data_dict)
    return final_data_set
                    
```

```python
file_url = 'GAC DPT_T9M_CDC_V0.2_20240729.xlsx'
rows_skip = 0
reject_value = 'Index'
data_set = parse_xls(file_url,rows_skip,reject_value)
print(data_set)
```

```python
import os
import zipfile
import xml.etree.ElementTree as ET

NS = {
    "x": "http://schemas.openxmlformats.org/spreadsheetml/2006/main"
}

def list_xlsx_sheets(folder_path):
    result = {}

    for file in os.listdir(folder_path):
        if not file.lower().endswith(".xlsx"):
            continue

        xlsx_path = os.path.join(folder_path, file)

        try:
            with zipfile.ZipFile(xlsx_path) as z:
                workbook_xml = z.read("xl/workbook.xml")

            root = ET.fromstring(workbook_xml)

            sheets = [
                sheet.attrib["name"]
                for sheet in root.findall(".//x:sheet", NS)
            ]

            result[file] = sheets

        except Exception as e:
            result[file] = {"error": str(e)}

    return result
```

```python
import zipfile
import xml.etree.ElementTree as ET

NS = {
    "x": "http://schemas.openxmlformats.org/spreadsheetml/2006/main",
    "r": "http://schemas.openxmlformats.org/officeDocument/2006/relationships"
}

def get_sheet_map(xlsx_path):
    with zipfile.ZipFile(xlsx_path) as z:
        workbook = ET.fromstring(z.read("xl/workbook.xml"))
        rels = ET.fromstring(z.read("xl/_rels/workbook.xml.rels"))

    rel_map = {
        rel.attrib["Id"]: rel.attrib["Target"]
        for rel in rels.findall(
            ".//{http://schemas.openxmlformats.org/package/2006/relationships}Relationship"
        )
    }

    sheet_map = {}

    for sheet in workbook.findall(".//x:sheet", NS):
        name = sheet.attrib["name"]
        rid = sheet.attrib["{http://schemas.openxmlformats.org/officeDocument/2006/relationships}id"]
        sheet_map[name] = "xl/" + rel_map[rid]

    return sheet_map
```

```python
xlsx = "data/GAC DPT_T9M_CDC_V0.2_20240729.xlsx"

sheet_map = get_sheet_map(xlsx)

print(sheet_map)
```
